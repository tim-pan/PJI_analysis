```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def draw_fig(path_csv):
    df = pd.read_csv(path_csv, index_col=False)
    df = df.iloc[:, 1:]
    cols = df.columns

    metric_cols = cols[-5:]
    para_set_cols = [item for item in cols if item not in set(metric_cols)]

    X = df[para_set_cols].to_numpy()
    Y = df[metric_cols].T.to_numpy()

    le = 1
    for _, y in zip(X, Y):
        le += 1
        plt.plot(y)

    plt.xticks(range(le), [str(item.tolist()) for item in X])
    plt.xlabel('parameter setup')
    plt.title(f'x cordinate means values of {para_set_cols}')
    plt.legend(['precision', 'recall', 'ROC', 'f1', 'acc'])
    plt.show()
    
    return df
```

```python
draw_fig('./CNN_result_smote.csv')
```

```python
draw_fig('./CNN_result_smote_pca.csv')
```
